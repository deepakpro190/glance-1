```python
from datasets import load_dataset

# Load dataset from local cache
dataset = load_dataset(
    "detection-datasets/fashionpedia",
    cache_dir="./data"
)

sample = dataset["train"][0]

print("=" * 80)
print("Sample Keys")
print("=" * 80)

print(sample.keys())

print()

print("=" * 80)
print("Image Information")
print("=" * 80)

print("Image ID :", sample["image_id"])
print("Width    :", sample["width"])
print("Height   :", sample["height"])

print()

print("=" * 80)
print("Objects")
print("=" * 80)

print(sample["objects"])

print()

print("=" * 80)
print("Image Type")
print("=" * 80)

print(type(sample["image"]))

print()

print("=" * 80)
print("Image")
print("=" * 80)

sample["image"].show()
```
